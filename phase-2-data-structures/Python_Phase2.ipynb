# Lesson 1: Lists

## 1.1 What Is a List?

A **list** is an ordered, mutable collection of items.
- **Ordered** — items stay in the position you put them.
- **Mutable** — you can change the list after creating it: add, remove, or modify items.

Think of a list like a **train**. Each carriage (element) has a fixed position (index). You can add carriages at the end, remove them from anywhere, and look inside any carriage by its number.

---

### 📊 [VISUAL] List Memory Layout

```
LIST:  scores = [85, 92, 78, 90, 65]

Memory layout (how Python stores it):

Index:    0       1       2       3       4
       +-------+-------+-------+-------+-------+
Value: |  85   |  92   |  78   |  90   |  65   |
       +-------+-------+-------+-------+-------+
Neg:    -5      -4      -3      -2      -1

scores[0]   -> 85          (first element)
scores[-1]  -> 65          (last element, negative index)
scores[1:3] -> [92, 78]    (slice: index 1 up to but not including 3)
```

## 1.3 Indexing and Slicing — Visual Deep Dive

Every element has two addresses: a **positive index** (left, starts at 0) and a **negative index** (right, starts at -1).

### 📊 [VISUAL] Indexing and Slicing

```
word = ['P', 'y', 't', 'h', 'o', 'n']

Positive index:  0    1    2    3    4    5
               +----+----+----+----+----+----+
               | P  | y  | t  | h  | o  | n  |
               +----+----+----+----+----+----+
Negative index: -6   -5   -4   -3   -2   -1

SLICING:  list[start : stop : step]
          stop is EXCLUSIVE (not included)

word[0:3]   -> ['P','y','t']              start=0, stop=3
word[2:]    -> ['t','h','o','n']          start=2, go to end
word[:4]    -> ['P','y','t','h']          from start, stop=4
word[::2]   -> ['P','t','o']              every 2nd element
word[::-1]  -> ['n','o','h','t','y','P']  reversed!
```

````python
nums = [10, 20, 30, 40, 50, 60]

print(nums[0])      # 10  — first
print(nums[-1])     # 60  — last
print(nums[2:5])    # [30, 40, 50]
print(nums[:3])     # [10, 20, 30]
print(nums[3:])     # [40, 50, 60]
print(nums[::2])    # [10, 30, 50]  — every 2nd
print(nums[::-1])   # [60, 50, 40, 30, 20, 10]  — reversed
````

## 1.4 Modifying Lists — All the Ways

### 📊 [VISUAL] Mutation Operations

```
append()   ->  adds ONE item to the END
insert()   ->  adds ONE item at a SPECIFIC position
extend()   ->  adds ALL items from another list to the END
remove()   ->  removes FIRST occurrence of a value
pop()      ->  removes and RETURNS item at index (default: last)
del        ->  deletes item at index or a slice
sort()     ->  sorts IN PLACE (modifies the original)
reverse()  ->  reverses IN PLACE
clear()    ->  removes ALL items
```

````python
fruits = ['apple', 'banana', 'cherry']

# append — add to end
fruits.append('date')
print("After append:", fruits)

# insert — at specific position
fruits.insert(1, 'avocado')
print("After insert:", fruits)

# extend — merge another list in
more = ['elderberry', 'fig']
fruits.extend(more)
print("After extend:", fruits)

# remove — by value (removes FIRST match only)
fruits.remove('banana')
print("After remove:", fruits)

# pop — by index, returns the removed value
removed = fruits.pop(0)
print(f"Popped: {removed}, List: {fruits}")

# sort — ascending by default
numbers = [3, 1, 4, 1, 5, 9, 2, 6]
numbers.sort()
print("Sorted asc:", numbers)

numbers.sort(reverse=True)
print("Sorted desc:", numbers)
````

## 1.5 Iterating Over Lists

> 🔵 **[AI/ML]** Training batches: `batch = dataset[0:32]`. Storing losses: `losses = []` then `losses.append(epoch_loss)` every epoch. Label lists: `labels = [0, 1, 0, 1, 1]`. NumPy arrays are lists with superpowers — same indexing, same slicing logic.

````python
scores = [85, 92, 78, 90, 65]

# Method 1: direct iteration (most common)
print("--- Direct ---")
for score in scores:
    print(score, end=' ')
print()

# Method 2: enumerate() — gives index AND value
print("--- Enumerate ---")
for index, score in enumerate(scores):
    print(f'Student {index+1}: {score}')

# Method 3: range + len — when you need index for modification
for i in range(len(scores)):
    scores[i] = scores[i] + 5    # add 5 bonus marks
print("After bonus:", scores)

# zip() — iterate two lists in parallel
names  = ['Alice', 'Bob', 'Carol']
grades = [88, 74, 95]
for name, grade in zip(names, grades):
    print(f'{name}: {grade}')
````

## 1.6 Common Mistakes — Lists

### ❌ [MISTAKE 1] IndexError — accessing an index that doesn't exist
### ❌ [MISTAKE 2] Confusing `sort()` (in-place) and `sorted()` (returns new list)
### ❌ [MISTAKE 3] The List Copy Trap — **the most dangerous beginner mistake**

### 📊 [VISUAL] The Copy Trap

```
a = [1, 2, 3]
b = a              # WRONG — b is NOT a copy!

Memory:
a ------>  [1, 2, 3]  <------ b
             ^
             Both 'a' and 'b' point to the SAME object in memory!
             Changing b changes a. Changing a changes b.

b = a.copy()       # CORRECT

Memory:
a ------>  [1, 2, 3]
b ------>  [1, 2, 3]   (separate object — changing b does NOT affect a)
```

````python
# MISTAKE 1: IndexError
nums = [1, 2, 3]
# print(nums[3])    # IndexError! Valid indices: 0, 1, 2
# print(nums[-4])   # IndexError! Valid negatives: -1, -2, -3
print("Valid access:", nums[2], nums[-1])

# MISTAKE 2: sort() vs sorted()
nums = [3, 1, 2]
result = sorted(nums)    # returns NEW sorted list, nums unchanged
print("sorted() result:", result)
print("original nums:", nums)   # unchanged!
nums.sort()                      # modifies nums in place, returns None
print("after .sort():", nums)

# MISTAKE 3: The Copy Trap
a = [1, 2, 3]
b = a               # WRONG — both point to same object!
b.append(4)
print("a after changing b:", a)  # [1,2,3,4] — a changed too!

# Correct ways to copy
a = [1, 2, 3]
b = a.copy()        # shallow copy  ✅
c = a[:]            # slice copy    ✅
d = list(a)         # constructor   ✅
d.append(99)
print("a after correct copy:", a)  # unchanged
````

# Lesson 2: List Comprehensions

## 2.1 What Is a List Comprehension?

A list comprehension is a **one-line way to create a new list** by transforming or filtering another sequence. It is the most Python-specific feature beginners must learn, and it is used constantly in data science and AI/ML.

### outputList = {expression(variable) for variable in inputList [if variable condition1][if variable condition2]}
- **expression**: Optional. expression to compute the members of the output List which satisfies the optional conditions
- **variable**: Required. a variable that represents the members of the input List.
- **inputList**: Required. Represents the input set.
- **condition1, condition2 etc**; : Optional. Filter conditions for the members of the output List.
### 📊 [VISUAL] List Comprehension Formula

```
FORMULA:

new_list = [expression  for item in iterable  if condition]
            ^^^^^^^^^^                         ^^^^^^^^^^^
            what to do     loop variable       optional filter
            with each item

EQUIVALENT LOOP (same result, more lines):

new_list = []
for item in iterable:
    if condition:           # optional
        new_list.append(expression)

READING TIP: read left to right like English:
[x*2  for x in range(5)]  =>  'x times 2, for each x in range(5)'
```

---

## 2.2 From Loop to Comprehension — Side by Side

````python
# TASK: square every number from 1 to 10

# Old way (loop)
squares_loop = []
for n in range(1, 11):
    squares_loop.append(n ** 2)
print("Loop:", squares_loop)

# Comprehension — same result, one line
squares = [n ** 2 for n in range(1, 11)]
print("Comp:", squares)

# TASK: keep only even numbers from 0-19
evens = [n for n in range(20) if n % 2 == 0]
print("Evens:", evens)

# TASK: uppercase all strings
words = ['hello', 'world', 'python']
upper = [word.upper() for word in words]
print("Upper:", upper)
````

## 2.3 Comprehensions with Conditions

> 🔵 **[AI/ML]** These patterns appear in almost every data preprocessing script:
> - Normalising: `normalised = [x/255 for x in pixel_values]`
> - Filtering: `valid = [row for row in dataset if row['age'] > 0]`
> - Extracting: `names = [row['name'] for row in records]`
> - Labels: `binary = [1 if label=='cat' else 0 for label in labels]`

> - **if ONLY (filter)** — condition at the END<br>
`[x for x in lst if condition]`

>- **if-else (transform)** — condition in the EXPRESSION at the START<br>
`[value_if_true if condition else value_if_false for x in lst]`

```

````python
marks = [82, 45, 91, 33, 67, 55, 29, 88]

# Filter: marks >= 50 only
passed = [m for m in marks if m >= 50]
print("Passed:", passed)

# Transform AND filter: normalise passing marks to 0–1 range
normalised = [m/100 for m in marks if m >= 50]
print("Normalised:", normalised)

# if-else inside comprehension (ternary style)
result = ['Pass' if m >= 50 else 'Fail' for m in marks]
print("Results:", result)

# AI/ML pattern: convert labels to binary
labels = ['cat', 'dog', 'cat', 'bird', 'dog']
binary = [1 if label == 'cat' else 0 for label in labels]
print("Binary labels:", binary)
````

# Lesson 3: Tuples

## 3.1 What Is a Tuple?

A tuple is an **ordered, immutable** collection. **Immutable** means once created, you cannot change it — no add, remove, or modify. This is not a limitation; it is a **feature**. Immutability signals: *"this data should never change."*

### 📊 [VISUAL] List vs Tuple

```
list:   scores = [85, 92, 78]    # square brackets
tuple:  point  = (10, 25)        # round brackets (parentheses)

LIST (mutable):                   TUPLE (immutable):
+-------+-------+-------+         +-------+-------+
|  85   |  92   |  78   |         |  10   |  25   |
+-------+-------+-------+         +-------+-------+
 Can be changed                    Locked. Cannot be modified.
 scores[0] = 99  # works           point[0] = 99  # TypeError!

WHEN TO USE TUPLE:
  -> Coordinates:    (latitude, longitude)
  -> RGB colour:     (255, 128, 0)
  -> Model shape:    (batch_size, height, width, channels)
  -> Return multiple values from a function
  -> Dictionary keys (lists cannot be dict keys, tuples can)
```

> 🔵 **[AI/ML]** Model input shape: `input_shape = (32, 224, 224, 3)`. Dataset split: `train, val, test = (0.7, 0.15, 0.15)`. Bounding boxes: `box = (x_min, y_min, x_max, y_max)`. `return loss, accuracy` — this returns a tuple. Tuple immutability **protects configuration data** from accidental modification during training.

---

## 3.2 Creating and Using Tuples

````python
# Creating tuples
point   = (10, 25)
rgb_red = (255, 0, 0)
person  = ('Arjun', 22, 'Ahmedabad')

# Single-element tuple — the COMMA is mandatory!
single    = (42,)   # THIS is a tuple
not_tuple = (42)    # THIS is just an int in parentheses
print(type(single))     # <class 'tuple'>
print(type(not_tuple))  # <class 'int'>

# Tuple packing and unpacking
coordinates = (28.6, 77.2)           # packing
latitude, longitude = coordinates    # unpacking
print(f"Lat: {latitude}, Lon: {longitude}")

# Unpacking in a loop
students = [('Alice', 88), ('Bob', 74), ('Carol', 95)]
for name, score in students:
    print(f'{name}: {score}')

# Swap variables using tuple unpacking — the Pythonic way
a, b = 10, 20
a, b = b, a
print(f"After swap: a={a}, b={b}")  # a=20, b=10
````

## 3.3 Named Tuples — Self-Documenting Tuples

````python
from collections import namedtuple

# Define a structure
Point   = namedtuple('Point', ['x', 'y'])
Student = namedtuple('Student', ['name', 'age', 'gpa'])

# Create instances
p1 = Point(10, 25)
s1 = Student('Riya', 22, 8.7)

# Access by name (readable) OR by index (also works)
print(p1.x)       # 10   <- readable
print(p1[0])      # 10   <- also works
print(s1.name)    # Riya
print(s1.gpa)     # 8.7

# Useful in ML: model shape
ModelShape = namedtuple('ModelShape', ['batch', 'height', 'width', 'channels'])
shape = ModelShape(32, 224, 224, 3)
print(f"Batch size: {shape.batch}, Image: {shape.height}x{shape.width}")
````

# Lesson 4: Dictionaries

## 4.1 What Is a Dictionary?

A dictionary stores data as **key-value pairs**. Instead of accessing elements by position (index 0, 1, 2…), you access them by a meaningful **label** called a key. This is the **most important data structure in Python for real-world programming**.

### 📊 [VISUAL] List vs Dictionary

```
LIST (index-based):             DICTIONARY (key-based):

student = ['Arjun',22,8.5]     student = {'name': 'Arjun',
                                            'age':  22,
student[0] -> 'Arjun'                      'gpa':  8.5}
student[1] -> 22
student[2] -> 8.5              student['name'] -> 'Arjun'
                               student['age']  -> 22
Problem: what is [1]?          student['gpa']  -> 8.5
You need to remember           CLEAR. Self-documenting.
the position.

DICTIONARY MEMORY MODEL:

Key ---------> Value
'name'   ----> 'Arjun'
'age'    ----> 22
'gpa'    ----> 8.5
'city'   ----> 'Ahmedabad'

Keys must be UNIQUE and IMMUTABLE (strings, ints, tuples).
Values can be ANYTHING — including lists, dicts, functions.
```

---

## 4.2 Creating and Accessing Dictionaries

````python
# Creating a dictionary
student = {
    'name':  'Arjun',
    'age':   22,
    'gpa':   8.5,
    'city':  'Ahmedabad',
    'marks': [85, 92, 78, 90]   # value can be a list!
}

# Access by key
print(student['name'])    # 'Arjun'
print(student['marks'])   # [85, 92, 78, 90]

# Safe access: .get() — returns None if key missing (NO crash)
print(student.get('name'))                    # 'Arjun'
print(student.get('email'))                   # None
print(student.get('email', 'Not provided'))   # 'Not provided'

# ALWAYS prefer .get() when you are not 100% sure the key exists
````

## 4.3 Modifying Dictionaries

````python
student = {'name': 'Arjun', 'age': 22}

# Add new key-value pairs
student['email'] = 'arjun@example.com'
student['gpa']   = 8.5
print("After add:", student)

# Modify existing value
student['age'] = 23
print("After modify:", student)

# Delete a key
del student['email']
print("After del:", student)

# pop — remove and return value
age = student.pop('age')
print(f"Popped age: {age}, Remaining: {student}")

# update — merge another dict in
extra = {'city': 'Mumbai', 'phone': '9876543210'}
student.update(extra)
print("After update:", student)
````

## 4.4 Dictionary Methods — The Essential Ones

### 📊 [VISUAL] Method Reference

```
METHOD              RETURNS                   USE CASE
-------             --------                  ---------
.keys()             dict_keys view            loop over all keys
.values()           dict_values view          loop over all values
.items()            dict_items view           loop over (key, value) pairs
.get(k, default)    value or default          safe access
.update(d)          None (modifies in place)  merge dicts
.pop(k)             removed value             remove and use
.copy()             new dict                  shallow copy
len(d)              int                       number of key-value pairs
k in d              True/False                check if key exists
```

> 🔵 **[AI/ML]** Hyperparameter configs: `params = {'lr':0.001, 'epochs':100}`. Label maps: `class_names = {0:'cat', 1:'dog', 2:'bird'}`. Metrics tracking: `history = {'loss':[], 'accuracy':[]}`. JSON APIs return dictionaries.

````python
config = {'lr': 0.01, 'epochs': 100, 'batch_size': 32}

# Loop over keys
print("Keys:", config.keys())

# Loop over values
print("Values:", list(config.values()))

# Loop over both — use .items() always for dict iteration
print("--- Key-Value pairs ---")
for key, value in config.items():
    print(f'  {key}: {value}')

# Check if key exists
if 'lr' in config:
    print(f"Learning rate: {config['lr']}")

if 'dropout' not in config:
    print("No dropout setting found")
````

# Lesson 5: Sets

## 5.1 What Is a Set?

A set is an **unordered collection of unique elements**. Two defining properties:
- **No duplicates** — automatically removed.
- **No guaranteed order** — you cannot say which element is "first".

Think of a set like a **bag where each item exists only once**.

### 📊 [VISUAL] Set Properties and Operations

```
LIST: [1, 2, 2, 3, 3, 3, 4]    # duplicates kept, ordered
SET:  {1, 2, 3, 4}              # duplicates removed, unordered

SET OPERATIONS (from mathematics):

A = {1, 2, 3, 4}     B = {3, 4, 5, 6}

Union (A | B):          {1, 2, 3, 4, 5, 6}   all from both
Intersection (A & B):   {3, 4}                in BOTH
Difference (A - B):     {1, 2}                in A but NOT in B
Sym. Diff (A ^ B):      {1, 2, 5, 6}          in one but NOT both

USE SETS WHEN:
  -> Membership check (in) is O(1) — much faster than lists
  -> Remove duplicates from a list
  -> Find common or unique elements between two collections
```

> 🔵 **[AI/ML]** Vocabulary: `vocab = set(all_words_in_corpus)`. Deduplication: `seen_ids = set()` then `if sample_id not in seen_ids`. Data leakage check: `set(train_ids) & set(test_ids)` — should be empty!

---

## 5.2 Creating and Using Sets

````python
# Creating sets
fruits  = {'apple', 'banana', 'cherry'}
numbers = {1, 2, 3, 4, 5}

# From a list — removes duplicates automatically
raw    = [1, 2, 2, 3, 3, 3, 4]
unique = set(raw)
print("Set from list:", unique)    # {1, 2, 3, 4}

# Empty set — MUST use set(), not {} (that creates an empty dict!)
empty_set  = set()    # ✅ correct
empty_dict = {}       # ❌ this is a dict!
print(type(empty_set))   # <class 'set'>
print(type(empty_dict))  # <class 'dict'>

# Add and remove
fruits.add('date')
fruits.remove('banana')   # raises KeyError if not found
fruits.discard('mango')   # safe — no error if not found
print("Fruits:", fruits)

# Membership check — O(1), extremely fast
print('apple' in fruits)   # True
print('grape' in fruits)   # False
````

## 5.3 Set Operations

````python
ai_skills  = {'Python', 'NumPy', 'Pandas', 'ML', 'Stats'}
web_skills = {'Python', 'HTML', 'CSS', 'JavaScript', 'SQL'}

# Union — all skills from both
all_skills = ai_skills | web_skills
print("Union:", all_skills)

# Intersection — skills needed for BOTH roles
common = ai_skills & web_skills
print("Common (intersection):", common)    # {'Python'}

# Difference — AI skills that are NOT web skills
ai_only = ai_skills - web_skills
print("AI only:", ai_only)

# Symmetric difference — in one but NOT both
unique_each = ai_skills ^ web_skills
print("Unique to each:", unique_each)

# Subset check
basics = {'Python', 'NumPy'}
print("Is basics a subset of ai_skills?", basics.issubset(ai_skills))  # True

# Data leakage check (AI/ML pattern)
train_ids = {1, 2, 3, 4, 5, 6, 7, 8}
test_ids  = {9, 10, 11, 12}
leak = train_ids & test_ids
print("Data leakage:", leak)   # set() — empty, no leakage!
````

# Lesson 6: Nested Data Structures

## 6.1 Why Nested Structures?

Real-world data is never flat. A dataset row is not just one value — it is a collection of properties. A model config is not one number — it is a tree of settings.

### 📊 [VISUAL] Flat vs Nested

```
FLAT DATA (limited):                NESTED DATA (real-world):

name   = 'Alice'                    student = {
age    = 22                             'name':   'Alice',
score1 = 88                             'age':    22,
score2 = 74                             'scores': [88, 74, 91],

## 6.2 List of Dictionaries — The Most Common Pattern

This is how every dataset is represented in Python before it goes into Pandas. It is also what JSON APIs return.

````python
# A dataset: list of records, each record is a dict
dataset = [
    {'name': 'Alice', 'age': 25, 'salary': 55000, 'dept': 'Engineering'},
    {'name': 'Bob',   'age': 30, 'salary': 62000, 'dept': 'Marketing'},
    {'name': 'Carol', 'age': 28, 'salary': 58000, 'dept': 'Engineering'},
    {'name': 'Dave',  'age': 35, 'salary': 71000, 'dept': 'Management'},
]

# Access one record
print(dataset[0])
print(dataset[0]['name'])   # 'Alice'

# Loop through all records
print("--- All records ---")
for person in dataset:
    print(f"  {person['name']}: ${person['salary']:,}")

# Filter using comprehension
engineers = [p for p in dataset if p['dept'] == 'Engineering']
print(f"Engineers: {[e['name'] for e in engineers]}")

# Compute average salary
avg_salary = sum(p['salary'] for p in dataset) / len(dataset)
print(f'Average salary: ${avg_salary:,.0f}')

# Sort by salary (descending)
sorted_data = sorted(dataset, key=lambda p: p['salary'], reverse=True)
print("Top earner:", sorted_data[0]['name'])
````

## 6.3 Dictionary of Lists — Configuration and Tracking

> 🔵 **[AI/ML]** When you call `model.fit()` in Keras/TensorFlow, the return value is **exactly this** — a dictionary of lists: `{'loss':[...], 'val_loss':[...], 'accuracy':[...]}`. Every ML engineer reads this to plot learning curves and diagnose underfitting vs overfitting.

````python
# ML training history — dict of lists
history = {
    'epoch':    [],
    'loss':     [],
    'val_loss': [],
    'accuracy': [],
}

# Simulate training loop
for epoch in range(1, 11):
    loss     = 1.0 / epoch
    val_loss = 1.1 / epoch
    acc      = 1 - loss

    history['epoch'].append(epoch)
    history['loss'].append(round(loss, 4))
    history['val_loss'].append(round(val_loss, 4))
    history['accuracy'].append(round(acc, 4))

# Print training summary
print("Epoch | Loss   | Val Loss | Accuracy")
print("-" * 40)
for i in range(len(history['epoch'])):
    e = history['epoch'][i]
    l = history['loss'][i]
    v = history['val_loss'][i]
    a = history['accuracy'][i]
    print(f"  {e:3d} | {l:.4f} |  {v:.4f}  | {a:.4f}")

# Best epoch
best_idx   = history['accuracy'].index(max(history['accuracy']))
best_epoch = history['epoch'][best_idx]
print(f"\nBest epoch: {best_epoch} (Accuracy: {max(history['accuracy']):.4f})")
````

# Lesson 7: Choosing the Right Data Structure

## 7.1 The Decision Framework

One of the most important engineering skills is knowing **which data structure to use for which problem**. Wrong choice = slow code, bugs, and confusion.

### 📊 [VISUAL] Decision Tree

```
START: What is your data?

Do you need a sequence in order?
|
YES -> Will it ever change (add/remove)?
       |
       YES -> Use LIST   [1, 2, 3]
       NO  -> Use TUPLE  (1, 2, 3)  -- signals 'this never changes'

Do you need to look things up by a label/name?
|
YES -> Use DICTIONARY  {'key': value}

Do you need unique items and set math?
|
YES -> Use SET  {1, 2, 3}
```

---

## 7.2 Quick-Reference Comparison

| Structure | Ordered | Mutable | Duplicates |
|-----------|---------|---------|------------|
| `list  [ ]` | ✅ Yes | ✅ Yes | ✅ Yes |
| `tuple ( )` | ✅ Yes | ❌ No  | ✅ Yes |
| `dict  {k:v}` | ✅ Yes (Py 3.7+) | ✅ Yes | Keys: ❌, Values: ✅ |
| `set   { }` | ❌ No  | ✅ Yes | ❌ No  |

---

## 7.3 AI/ML Use Case Mapping

| Use Case | Best Structure | Why |
|----------|---------------|-----|
| Training data samples | list of dicts | Ordered, mutable, labelled fields |
| Model input shape | tuple | Fixed, immutable, e.g. `(32,224,224,3)` |
| Hyperparameter config | dict | Named parameters, easy to update |
| Unique vocabulary (NLP) | set | Fast lookup, no duplicates |
| Pixel values of an image | list / NumPy | Ordered, large, needs math |
| Class label map | dict | `{0:'cat', 1:'dog'}` readable lookup |
| Dataset split ratios | tuple | `(0.7, 0.15, 0.15)` should never change |
| Tracking processed IDs | set | O(1) membership, no duplicates |

# Lesson 8: Essential Built-in Functions for Data

## 8.1 Functions That Work on All Sequences

````python
data = [45, 82, 67, 29, 91, 55, 73]

print("len:   ", len(data))              # 7
print("sum:   ", sum(data))              # 442
print("min:   ", min(data))              # 29
print("max:   ", max(data))              # 91
print("sorted:", sorted(data))           # ascending
print("sorted desc:", sorted(data, reverse=True))

# enumerate — index + value
print("--- enumerate ---")
for i, val in enumerate(data):
    print(f'  [{i}] = {val}')

# zip — pair up two lists
names  = ['Alice', 'Bob', 'Carol']
scores = [88, 74, 95]
pairs  = list(zip(names, scores))
print("\nzipped:", pairs)

# map — apply a function to every element
doubled = list(map(lambda x: x * 2, data))
print("doubled:", doubled)

# filter — keep elements that pass a test
high = list(filter(lambda x: x >= 70, data))
print("high scores:", high)

# any / all
print("\nany > 90:", any(x > 90 for x in data))    # True (91)
print("all > 20:", all(x > 20 for x in data))       # True
````

## 8.2 The `key=` Parameter — Sorting Complex Data

````python
students = [
    {'name': 'Carol', 'gpa': 9.1},
    {'name': 'Alice', 'gpa': 8.5},
    {'name': 'Bob',   'gpa': 7.8},
]

# Sort by gpa ascending
by_gpa = sorted(students, key=lambda s: s['gpa'])
print("Ascending GPA:", [s['name'] for s in by_gpa])

# Sort by gpa descending
top_first = sorted(students, key=lambda s: s['gpa'], reverse=True)
print("Top first:", top_first[0]['name'])   # Carol

# Sort by name alphabetically
by_name = sorted(students, key=lambda s: s['name'])
print("Alphabetical:", [s['name'] for s in by_name])

# Find student with highest gpa using max()
topper = max(students, key=lambda s: s['gpa'])
print(f"Topper: {topper['name']} (GPA: {topper['gpa']})")
````
