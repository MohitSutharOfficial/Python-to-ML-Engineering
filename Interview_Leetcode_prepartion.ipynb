# 🎯 FAANG Interview Preparation Strategy

## How to Approach Any DSA Problem

```
STEP 1 — UNDERSTAND (2 min)
  Read twice. Identify:
  - Input type, size, constraints
  - Output format
  - Edge cases: empty input, single element, negatives, duplicates

STEP 2 — EXAMPLES (2 min)
  Walk through given examples manually
  Create your own edge case example

STEP 3 — BRUTE FORCE (2 min)
  State the naive O(n²) or O(n³) solution
  "The brute force would be... which is O(n²)"
  Never code the brute force — just state it

STEP 4 — OPTIMIZE (5 min)
  Identify the bottleneck
  Ask: "Which data structure makes this O(1)?"
  Ask: "Can I precompute something?"
  Ask: "Is there a sliding window here?"

STEP 5 — PSEUDOCODE (3 min)
  Write 5 lines of pseudocode
  Verify logic with your example
  THEN start coding

STEP 6 — CODE (10 min)
  Write clean, readable code
  Use meaningful variable names
  Add a comment for non-obvious logic

STEP 7 — TEST (3 min)
  Walk through your code with example input
  Test edge cases: [], [1], negative numbers
  State the time and space complexity
```

## Debugging Strategy

```
WHEN YOUR CODE FAILS:
  1. Print the input at the very first line
  2. Print the state after EVERY major operation
  3. Trace manually on the smallest failing case
  4. Check: off-by-one in range? Inclusive vs exclusive?
  5. Check: mutating a list while iterating?
  6. Check: shallow copy vs deep copy?
  7. Check: integer overflow? (Usually not in Python)
  8. Check: base case in recursion?

COMMON PYTHON BUGS IN INTERVIEWS:
  ❌ for i in range(len(arr)-1)  → misses last element sometimes
  ❌ if node.left or node.right  → should be if node.left: and if node.right:
  ❌ list.sort() instead of sorted(list) → modifies original
  ❌ x = arr[:]  → only shallow copy (OK for 1D, bad for 2D)
  ❌ Using [] as default arg → def func(lst=[]) → shared across calls
```
